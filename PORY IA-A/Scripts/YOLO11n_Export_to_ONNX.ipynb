# 🚀 Conversión de YOLO11 a ONNX para .NET MAUI

Este notebook documenta el proceso técnico para transformar el modelo original de **Ultralytics** al formato **ONNX** (Open Neural Network Exchange).

### 🎯 Objetivo
Optimizar el modelo **YOLO11n** para que pueda ser ejecutado de manera eficiente en dispositivos móviles (Android) y escritorio (Windows) utilizando el motor de inferencia de **Microsoft.ML.OnnxRuntime** con **.NET 10**.

---

## 🛠️ Paso 1: Preparación del Entorno
En esta etapa instalamos las librerías necesarias. `ultralytics` es el framework principal que nos permite manipular los modelos YOLO de última generación.

```python
# Instalar Ultralytics
!pip install ultralytics

from ultralytics import YOLO
import os
```

---

## 📦 Paso 2: Selección del Modelo Base
Cargamos el modelo **YOLO11n (Nano)**. Hemos seleccionado esta variante específica porque es la más ligera de la familia YOLO11.

Su arquitectura está diseñada para maximizar la velocidad de detección en procesadores móviles sin sacrificar drásticamente la precisión, lo que la hace ideal para aplicaciones de tiempo real en **.NET MAUI**.

```python
# Usamos YOLO11n porque es el modelo más pequeño y eficiente
model = YOLO('yolo11n.pt')
```

---

## ⚙️ Paso 3: Exportación y Optimización a ONNX
Este es el paso fundamental para la integración con **.NET**. Configuramos la exportación con parámetros específicos para garantizar la estabilidad en el motor de inferencia:

*   **`imgsz=[640, 640]`**: Definimos una resolución fija. Esto evita errores de tensores dinámicos en dispositivos móviles y asegura un rendimiento constante.
*   **`opset=17`**: Utilizamos una versión moderna y robusta de ONNX, totalmente compatible con las librerías de **Microsoft.ML.OnnxRuntime**.
*   **`simplify=True`**: Aplicamos una optimización al grafo del modelo para eliminar operaciones redundantes, resultando en un archivo más liviano.

```python
# Exportar al formato ONNX
model.export(
    format="onnx",      # Formato de destino
    imgsz=[640, 640],   # Forzamos el tamaño de entrada para evitar problemas de tensores dinámicos
    opset=17,           # Versión de operaciones ONNX (17 es muy estable para .NET)
    simplify=True       # Optimiza el grafo del modelo eliminando nodos innecesarios
)
```

---

## 🏁 ¡Listo para usar!
Una vez finalizada la ejecución, el archivo optimizado se encontrará en la carpeta del modelo con el nombre `yolo11n.onnx`.

   `Resources/Raw/yolo11n.onnx`
